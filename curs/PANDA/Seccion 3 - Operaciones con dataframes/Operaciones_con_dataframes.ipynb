## Lección 2 - Datos missings

```python
import pandas as pd
import numpy as np

from pathlib import Path

data_path = Path('./data')
```

```python
data = pd.read_csv(data_path / 'house.csv')
data.head(2)
```

```python
data.isna().sum()
```

Hemos preparado un pequeño código para añadir algunos valores missings.

```python
_ = data.sample(frac = 0.05)
data['price'] = pd.Series([np.nan if i in _.index.values else v for i,v in zip(np.arange(len(data)), data['price'])])
_ = data.sample(frac = 0.05)
data['bedrooms'] = pd.Series([np.nan if i in _.index.values else v for i,v in zip(np.arange(len(data)), data['bedrooms'])])
```

```python
data.isna().sum()
```

```python
data['bedrooms'].value_counts(dropna = False)
```

#### con notna() podemos filtrar los valores missing

```python
data['price']
```

```python
data[data['price'].notna()]['price']
```

Otra opción es con la funcion query, y aprovechando que nan == nan = False

```python
np.nan == np.nan
```

```python
data.query('bedrooms == bedrooms').shape
```

#### Calculos en columnas con datos missings

```python
data['bedrooms'].mean(skipna = True)
```

```python
data['bedrooms'].mean(skipna = False)
```

```python
data['bedrooms'].min(skipna = True)
```

```python
data['bedrooms'].min(skipna = False)
```

#### Fillna para rellenar los valores missings

```python
_ = data['bedrooms'].fillna(value = -1)
_
```

Tambien podemos rellenar con el valor anterior o posterior.

```python
_ = data['bedrooms'].fillna(method='backfill')
pd.Series(_).value_counts()
```

Es posible que queramos simplemente excluir aquellas columnas o filas que tengan algun NaN. Para hacer esto, use dropna ():

```python
# axis = 'index' elimina todas las filas con algun nan
data.dropna(axis = 'index', how = 'any').shape
```

```python
# axis = 'columns' elimina todas las columnas con algun nan
data.dropna(axis = 'columns', how = 'any').shape
```

## Lección 3 - Operaciones entre dataframes

```python
import pandas as pd
import numpy as np

from pathlib import Path

data_path = Path('./data')
```

```python
data = pd.read_csv(data_path / 'house.csv')
data.head(2)
```

Es habitual tener que trabajar con varios dataframes. A veces, necesitamos unir o concatenar dichos dataframes. Lo hacemos de la siguiente forma:

```python
df1 = data.iloc[0:10]
df1.head()
```

```python
df2 = data.iloc[-10:]
df2.head()
```

¿Cómo genero un único DF con los datos de ambos? Lo hago con el método **concat**

```python
df = pd.concat([df1, df2], axis = 0)
print(df.shape)
df.head()
```

```python
df = pd.concat([df1, df2], axis = 0, keys = ['head', 'tail'])
print(df.shape)
df.head()
```

```python
df.loc['tail'].head()
```

También podemos realizar la concatenación con el método append. 

```python
df1.append(df2, ignore_index=True).head()
```

¿Qué ocurre si es el caso contrario? Es decir, si tengo el 2 dataframes con los mismos índices, pero distintas columnas. También puedo concatenarlos.

```python
df1 = data.loc[0:10, ['price', 'bedrooms']]
df1
```

```python
df2 = data.loc[0:10, ['city', 'statezip', 'country']]
df2
```

```python
df = pd.concat([df1, df2], axis = 1)
df
```

#### Join 2 DF, de manera similar a dos tablas de BBDD

```python
df1 = data.loc[0:10, ['city', 'price', 'bedrooms']]
df1.head()
```

```python
df2 = data.loc[0:10, ['city', 'statezip', 'country']]
df2.head()
```

```python
df = df1.merge(df2, how = 'left', left_on = 'city', right_on = 'city')
df.head()
```

En este ejemplo, hemos cruzado por una sola columna, pero podemos hacerlo por varias: left_on = [columna1, columna2], right_on = [columna1, columna2]

Join tiene un funcionamiento similar, pero lo solemos hacer por indice.

```python
df = df1.join(df2, lsuffix = '_x', rsuffix = '_y')
df.head()
```

## Lección 4 - Tablas pivotantes

Las tablas pivotantes nos sirven para situar valores de filas como columnas de nuestro DF.

```python
import pandas as pd
import numpy as np

from pathlib import Path

data_path = Path('./data')
```

```python
df = pd.read_csv(data_path / 'pivot_data.csv')
df.head(10)
```

Yo puedo consultar los valores de variable = "A" de una forma muy facil:

```python
df[df['variable'] == "A"]
```

Pero supongamos que deseamos realizar operaciones de series de tiempo con las variables. Una mejor representación sería donde las columnas son las variables únicas y un índice de fechas identifica las observaciones individuales. Para ello, usamos la pivot table

```python
_ = df.pivot(index='date', columns='variable', values='value')
_
```

```python
_ = df.pivot(index='variable2', columns='variable', values='date')
_
```

```python
_.shape
```

```python
_.index
```

Además, podemos realizar la agrupación por dos o más variables:

```python
_ = df.pivot(index=['date', 'variable'], columns='variable2', values='value')
_
```

```python
_.index
```

Estrechamente relacionados con el método pivot () están los métodos stack() y unstack()disponibles tanto en Series como DataFrame.

- Stack(): Disminuye el nº de columnas. "Deshace" el pivot table.
- Unstack(): Mueve uno de los indices a nivel columna.

```python
stack = _.stack().to_frame()
stack
```

Podemos ver como la "variable2", que antes en el DF anterior estaba formada por 3 columnas (una por cada valor), ahora solo cuenta como una, incrementando el nº de filas.

```python
stack.unstack(level = 0)
```

```python
stack.unstack(level = 1)
```

También podemos aplicar el unstack a más de una variable

```python
stack.unstack(level=['date','variable'])
```

## Lección 5 - Agrupaciones

"Group by" es un proceso que involucra uno o más de los siguientes pasos: 
- Dividir los datos en grupos según algunos criterios. 
- Aplicar una función a cada grupo de forma independiente. 
- Combinar los resultados en una estructura de datos.

Este proceso de agrupación y agregación es muy similar al que efectuamos en SQL.

```python
import pandas as pd
import numpy as np

from pathlib import Path

data_path = Path('./data')
```

```python
data = pd.read_csv(data_path / 'house.csv')
data.head(2)
```

```python
g = data.groupby('condition')
g
```

```python
g['price']
```

```python
g['price'].mean()
```

También, puedo realizar las agrupaciones por más de una columna, de manera que genero un multiIndex. 

```python
g = data.groupby(['condition','waterfront'])['price'].mean()
g
```

```python
g.index
```

Puedo convertir esa serie en un DF con el método to_frame() o reset_index()

```python
g.reset_index()
```

```python
dfg = g.to_frame()
dfg
```

En lugar de agrupar por una columna, también es posible agrupar por un valor del indice.

```python
dfg.groupby(level=1)['price'].mean()
```

```python
dfg.groupby(level=1).mean()
```

Hasta ahora, solo hemos usado una variable para la agregación, pero podemos aplicar la operación a más de una.

```python
_ = data[['price','bathrooms','condition','waterfront']]
_
```

```python
_.groupby(['condition','waterfront']).mean()
```

#### Agregaciones. 

Al realizar una agrupacion de distintas variables, podemos querer aplicar operaciones distintas a cada una de ellas. Para ello, usamos la agregación

```python
data.groupby('condition').agg({
    'price' : 'mean',
    'bathrooms' : 'median',
    'waterfront' : 'max',
    'floors' : 'min'
})
```

```python
data.groupby('condition').agg({
    'price' : np.mean,
    'bathrooms' : 'median',
    'waterfront' : np.max,
    'floors' : 'min'
})
```

También puedo aplicar distintas operaciones a la misma columna agregada

```python
agrupacion = data.groupby('condition').agg({
    'price' : [np.mean, np.min, np.max],
    'bathrooms' : 'median',
    'waterfront' : np.max,
    'floors' : ['min', 'max']
})
agrupacion
```

```python
agrupacion['price']['amax']
```

```python
agrupacion.stack(level=0)
```

También podemos usar el método apply para personalizar la funcion que deseemos.

```python
def media_mayores_1millon(x):
    return np.mean(x[x > 1_000_000])
```

```python
data[['price','condition']].groupby('condition')['price'].apply(media_mayores_1millon)
```
