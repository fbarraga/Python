La **asimetría (skewness)** y la **curtosis (peakedness)** son dos mesures estadístiques que serveixen para descriure la forma d'una distribució sense haver de fer la seva representació gráfica. En concret, la **asimetría (skewness)** indica el grau de simetria (o asimetria) d'una distribución, mentres que la curtosis indica el grau de concentració d'una distribució al voltant de la seva mitja.
Podem conèixer el skewness i la kurtosis de la gràfica utititzant els mètodos skew y kurt

```python
df_train['SalePrice'].skew()
```

```python
df_train['SalePrice'].kurt()
```

Declarem una variable `corrmat=df.train.corr()` amb totes les correlacions entre variables. Especificcant el numeric_only li diem que només volem les dades numèriques.

<h3>4.2- Matriu de correlació amb números</h3>


A vegades no és fàcil veure les correlacions amb un *heatmap* i es millor utilitzar números. Anem a fer lo mateix però amb números i veurem clarament quin camp es correlaciona més.


Comencem declarant K=10

```python
k=10
```

Ara agafem les deu variables que es correlacionen més amb la variable 'SalePrice'


```python
cols=corrmat.nlargest(k,"SalePrice")["SalePrice"].index
cols
```

A continuació anem a cm=np.corrcoef(df_train[cols],values,T)

```python
cm = np.corrcoef(df_train[cols].values.T)
```

```python
cm
```

```python
sns.set(font_scale=1.25)
```

Ara solapem el heatmap amb les dades numériques


```python
hm=sns.heatmap(cm,cbar=True,annot=True,square=True,fmt=".2f",annot_kws={"size":10,},yticklabels=cols.values,xticklabels=cols.values)
plt.show
```

Veiem que els valors més alts els tenim amb més espai de garatge (0,88) i  més cotxes al garatge (0,88)
