Ahora vamos pidiendo una foto - hagamos una gráfica de distplot

Veremos que:
<li>Se desvia de la distrbución normal</li>
<li>Tiene skew positivo (oblicuidad)</li>

Vamos viendo que sucede con las amigas de 'Precio Venta' Dibujemos un scatterplot de precioventa vs areavivienda. Nota como vamos a reducir el dataframe a solo 2 columnas para no tener que usar todo. Primero declara una variable var que contenga el literal del área visible

```python
var = 'GrLivArea'
```

Luego camos a usar pd.concat para crear una tabla de solo 2 columnas con area vivible(var) y precio de venta que se llame data

```python
data = pd.concat([df_train['SalePrice'],df_train[var]],axis=1)
```

```python
data.head(20)
```

y ahora vamos a terminar con un data.plot.scatter para crear una grafica de scatterplot

```python
data.plot.scatter(x=var,y='SalePrice',ylim=(0,800000))
```

Ahora que pasa con "SalePrice" y "AreaSotano". Son amigas pero su relacion no es lineasl. Sin mencionar que a veces el AreaSotano se va en su onda y evita el 'SalePrice'

```python
var='TotalBsmtSF'
data=pd.concat([df_train['SalePrice'],df_train[var]],axis=1)
data.plot.scatter(x=var,y='SalePrice',ylim=(0,800000))
```

<h3>Relaciones categóricas</h3>

Vamos ahora a armar una gráfica de boxplot para verificar el efecto de la calidad sobre el precio de ventas.
Comenzamos igual, declara una variable var='CalidadTotal' y una data donde concatenes el precio de ventas con la var

```python
var='OverallQual'
```

```python
data=pd.concat([df_train['SalePrice'],df_train[var]],axis=1)
data.head(5)
```

Ahora viene lo dificil, en dos variables f y ax, vas a guardar el resultado de un plt.subplots- donde se define la nueva figura donde existira la grafica que se va a crear. F y ax es la figura y los ejes que devuelve la instrucción


```python
f, ax = plt.subplots(figsize=(8,6))
fig = sns.boxplot(x=var,y="SalePrice",data = data)
fig.axis(ymin=0,ymax=800000)
```

por ultimo, definimos fig como una varibles donde haremos un sns.bloxplot(para guardar la grafica)

```python
f,ax = plt.subplots(figsize=(20,8))
fig=sns.boxplot(x=var,y="SalePrice",data=data)
fig.axis(ymin=0,ymax=800000)
plt.xticks(rotation=90) #rotar los titulos deel eje
```
