# Red Multicapa (Perceptrón Multicapa o MLP)

Una red multicapa nos sirve para resolver problemas más complejos. En las redes multicapa las salidas de algunas neuronas se convierten en la entrada de las siguientes neuronas, permitiendonos tener más que una clasificación o regresión meramente lineales. En este ejemplo crearemos una red multicapa con varias neuronas dentro de cada capa.

Para este ejemplo vamos a utilizar la ecuación de Rosenbrock, la ecuación de Rosenbrock fue propuesta por él con el fin de evaluar un método automático para minimizar una función. En su trabajo Rosenbrock utilizó una función de dos variables, que es la que nosotros utilizaremos en este ejemplo.

Ecuación de Rosenbrock:<br>
$f(x,y)=100\left(x^{2}-y\right)^{2}+\left(x-1\right)^{2}$

Aunque, esta ecuación no está límitada a dos variables, de hecho, puede utilizarse para funciones de n variables generalizandola como se muestra a continuación.

Ecuación de Rosenbrock generalizada:<br>
$f(x)=\sum_{i=1}^{n-1}\left[100\left(x_{i}^{2}-x_{i+1}\right)^{2}+\left(x_{i}-1\right)^{2}\right]$

Importamos librerías de utilidad.

```python
import numpy as np
import matplotlib.pyplot as plt 
import neurolab as nl
```

Vamos a plantear [-5, 10] como el espacio en el que trabajaremos esta función. Para los datos de entramiento crearemos un set de 50 datos en este intervalo.

```python
lim_inf = -5
lim_sup = 10
ejex = ejey = np.linspace(lim_inf, lim_sup, 50)
```

Ya con nuestro espacio creado procedemos a definir la función **Rosenbrock**. La función la vamos a trabajar de manera vectorial, es decir, en lugar de evaluar un punto a la vez vamos a evaluar todo el vector con el fin de ahorrarnos tiempo computacional. **Rosenbrock** va a recibir dos variables **x**, **y**, ambas vectores que nos facilitarán las operaciones algebraicas. Con estos vectores, definimos la función y la guardamos en **z** (al elevar al cuadrado un vector de *numpy* se realiza elemento por elemento, no es la multiplicación de matrices que se realiza en el álgebra lineal).

Finalmente vamos a devolver el resultado en **z**.

```python
def rosenbrock(x, y):
    z = 100*(x**2 - y)**2 + (x-1)**2
    return z
```

Ahora, vamos a crear las matrices que forman las parejas ordenadas que le mandaremos a la función para crear una superficie utilizando **meshgrid**, después llamamos a la función **rosebrock** mandándole las matrices que creamos y guardando la matriz que nos retorna en **z**. Finalmente procedemos a graficar.

```python
# creamos las parejas ordenadas
x, y = np.meshgrid(ejex, ejey)
# evaluamos la función
z = rosenbrock(x, y)
# se procede a graficar
plt.gca(projection='3d').plot_surface(x, y, z, cmap='jet')
plt.title("Función de Rosenbrock")
plt.show()
```

Ya que tenemos los valores que toma la función y los valores que nos retorna, hay que preparar esos datos para entrenar nuestra red neuronal.

Definimos el rango de que tienen nuestros datos de entrada que es [-5, 10], para ambas variables, y normalizamos los datos de **z** porque la función nos devuelve valores muy grandes (algunos mayores a 800,000).

```python
# establecemos las dimensiones de nuestras entradas
dim = [lim_inf, lim_sup]
# normalizamos los datos
z = z/np.linalg.norm(z)
```

Luego procedemos a transponer los datos (requisito para utilizar crear nuestra red con *neurolab*).

```python
# transponemos nuestros datos
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
z = z.reshape(-1, 1)
```

Y finalmente vamos a concatenar **x** y **y**, que son nuestros valores de entrada, ya que tienen que estar en una sola matriz cuando los pasemos a nuestra red para entrenarla. Este vector tiene las mismas dimensiones que establecimos anteriormente para ambas entradas. **z** lo dejamos igual porque solo queremos una neurona de salida.

```python
X = np.concatenate((x, y), axis=1)
X
```

### SI IMPORTAMOS LA RED YA CREADA SE OMITE EL SIGUIENTE PASO (ENTRENAMIENTO) Y SE PROCEDE A REALIZAR LA PRUEBA

```python
# IMPORTAR LA RED YA CREADA
# net = nl.load("Rosenbrock.net")
```

Ahora vamos a crear nuestra red, la crearemos con 3 capas, 10 neuronas en la primera, 7 en la segunda y 3 en la tercera, además solo tendremos una neurona de salida, esta red multicapa la creamos con la función **newff** que recibe como parámetros las dimensiones de las neuronas de entrada y como lista las neuronas que habrá en cada capa exceptuando la capa de entrada.

Vamos a establecer el error de la red a MAE (*Mean Absolute Error*) y procedemos a entrenarla. Le mandamos las entradas que están en **X** y las salidas deseadas que están en **z**, la limitamos a 200 épocas y a un error mínimo de 0.001.

Después graficamos el progreso del error.

```python
# creamos el modelo
net = nl.net.newff([dim, dim], [10, 7, 3, 1])

# cambiamos el error
net.errorf = nl.error.MAE()

# entrenamos la red
error = net.train(X, z, epochs=200, show=20, goal=0.001)

# graficamos el error
plt.figure()
plt.plot(error)
plt.xlabel("Número de épocas")
plt.ylabel("Error")
plt.title("Progreso del error")
```

Vamos a generar datos de prueba para ver como se comporta nuestra red, el conjunto de datos de prueba que generaremos es un conjunto más denso que el que utilizamos para entrenar la red, pues este consta de 100 elementos. Una vez preprocesados los datos los evaluamos con la red con la función **sim**, y los graficamos para ver los datos reales vs los datos predichos. Después de preparar los datos para evaluarlos con nuestra red, tenemos que reacomodarlos a su forma original para poder graficar sin problemas, la forma original la podemos tomar de **z** y la guardamos en **forma** para reordenarlos después de evaluarlos.

```python
# generamos datos de prueba
ejex = ejey = np.linspace(lim_inf, lim_sup, 100)
# creamos las parejas ordenadas
x, y = np.meshgrid(ejex, ejey)
# evaluamos la función con los datos de prueba
z = rosenbrock(x, y)
# guardamos la forma de los datos
forma = z.shape
# normalizamos los datos
z = z/np.linalg.norm(z)
# transponemos nuestros datos
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
# concatenamos los datos de entrada
X = np.concatenate((x, y), axis=1)
# evaluamos la red
z_pred = net.sim(X)
# reacomodamos los datos para graficarlos
x = x.reshape(forma)
y = y.reshape(forma)
z_pred = z_pred.reshape(forma)
```

Y procedemos a graficar.

```python
plt.gca(projection='3d').plot_surface(x, y, z, cmap='jet')
plt.title("Función de Rosenbrock Real")
plt.show()
```

```python
plt.gca(projection='3d').plot_surface(x, y, z_pred, cmap='jet')
plt.title("Datos Predichos")
plt.show()
```
